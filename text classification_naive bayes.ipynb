```python
import numpy as np
import pandas as pd
import operator
from sklearn import model_selection
import re,string
import os,sys
import matplotlib.pyplot as plt
```

```python
stop_words=["a","about","above","after","again","against","all","am","an","and","any","are","as","at","be","because","been","before","being","below","between","both","but",
"by","could","did","do","does","doing","down","during","each","few","for","from","further","had","has","have","having","he","he'd","he'll","he's","her",
"here","here's","hers","herself","him","himself","his","how","how's","i","i'd","i'll","i'm","i've","if","in","into","is","it","it's","its","itself","let's","me",
"more","most","my","myself","nor","of","on","once","only","or","other","ought","our","ours","ourselves","out","over","own","same","she",
"she'd","she'll","she's","should","so","some","such","than","that","that's","the","their","theirs","them","themselves","then","there","there's",
"these","they","they'd","they'll","they're","they've","this","those","through","to","too","under","until","up","very","was","we","we'd",
"we'll","we're","we've","were","what","what's","when","when's","where","where's","which","while","who","who's","whom","why","why's","with",
"would","you","you'd","you'll","you're","you've","your","yours","yourself","yourselves"]
```

```python
x=[]   #x is the list where 1st element is name of document and second element is text in document
y=[]   # y is the category
for category in os.listdir("C:/Users/admin/20_newsgroups"):
    for document in os.listdir("C:/Users/admin/20_newsgroups/"+category):
        with open("C:/Users/admin/20_newsgroups/"+category+'/'+document, "r") as f:
            x.append((document,f.read()))
            y.append(category)
```

```python
xtrain,xtest,ytrain,ytest=model_selection.train_test_split(x,y)
```

```python
len(x)
```

```python
dic={}    #words with coreesonding frequency
for i in range (len(xtrain)):
    word=xtrain[i][1].lower()
    stripped=re.split(r'\W+',word)
    for s in stripped:
        if not (s.isalpha()) or s in stop_words or len(s)<=2:
            continue
        if s in dic:
            dic[s]+=1
        else:
            dic[s]=1
    
```

```python
sorted_dic = sorted(dic.items(), key=operator.itemgetter(1),reverse=True)
sorted_dic
```

```python
features=sorted_dic
ans1=[]
ans2=[]
for i in range (len(features)):
    ans1.append(i)
    ans2.append(features[i][1])
```

```python
plt.plot(ans1,ans2)
plt.axis([0,8000,1,1000])
plt.grid()
plt.show
```

```python
ans1=[features[i][0] for i in range (2000)]
```

```python
ans1
```

```python
x_train=np.zeros([len(xtrain),len(ans1)],int)
for i in range(len(xtrain)):
    words=xtrain[i][1].lower()
    word=re.split(r'\W+',words)
    for j in word:
            if j in ans1:
                 x_train[i][ans1.index(j)]+=1
        
```

```python
x_test=np.zeros([len(xtest),len(ans1)],int)
for i in range(len(xtest)):
    words=xtest[i][1].lower()
    word=re.split(r'\W+',words)
    for j in word:
        if j in ans1:
            x_test[i][ans1.index(j)]+=1
```

```python
from sklearn.naive_bayes import MultinomialNB
```

```python
clf=MultinomialNB()
```

```python
clf.fit(x_train,ytrain)
```

```python
clf.score(x_test,ytest)
```
